```python


# Sample text (you can change this)
text = """Monolithic architecture is simple.
Microservices architecture is scalable.
Each service works independently."""

def count_spaces_lines(text):
    lines = [line for line in text.split('\n') if line.strip()]  # non-empty lines
    spaces = text.count(' ')
    return len(lines), spaces

# Run
print(count_spaces_lines(text))
```
