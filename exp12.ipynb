```python


def first_follow(grammar, start_symbol):
    first = {nt: set() for nt in grammar}
    follow = {nt: set() for nt in grammar}

    # FOLLOW of start symbol contains $
    follow[start_symbol].add('$')

    # ---------- FIRST (basic logic) ----------
    for nt in grammar:
        for production in grammar[nt]:
            first[nt].add(production[0])  # take first symbol (simplified)

    # ---------- FOLLOW (basic logic) ----------
    for nt in grammar:
        for production in grammar[nt]:
            for symbol in production:
                if symbol in grammar:  # non-terminal
                    follow[symbol].add(nt)

    return first, follow


# Example Grammar
grammar = {
    'S': ['AB'],
    'A': ['a', 'ε'],
    'B': ['b']
}

first, follow = first_follow(grammar, 'S')

print("FIRST:", first)
print("FOLLOW:", follow)
```
